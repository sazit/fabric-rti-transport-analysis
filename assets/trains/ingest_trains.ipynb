## Real-Time Sydney Trains Position Monitor

This notebook connects to the Transport NSW **Sydney Trains v2 API** and feeds incoming position data into an Eventstream.

The v1 endpoint was deprecated 27 May 2025. This uses the v2 GTFS Realtime feed which provides richer data including train line, direction, stop, and movement status.

```python
import requests
import pandas as pd
from google.transit import gtfs_realtime_pb2
from azure.servicebus import ServiceBusClient, ServiceBusMessage
import json
import time
from datetime import datetime
import pytz
```

```python
# Transport NSW API key
myapikey = ""

# Eventstream Custom Endpoint connection string - get from Stream_Train_Loc
myconnectionstring = ""
```

```python
def get_train_positions(api_key):
    """Fetch real-time train positions from Transport NSW Sydney Trains v2 API."""
    url = "https://api.transport.nsw.gov.au/v2/gtfs/vehiclepos/sydneytrains"
    headers = {
        "Authorization": f"apikey {api_key}",
        "Accept": "application/x-google-protobuf"
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()

    feed = gtfs_realtime_pb2.FeedMessage()
    feed.ParseFromString(response.content)

    trains = []
    sydney_tz = pytz.timezone("Australia/Sydney")

    for entity in feed.entity:
        vehicle = entity.vehicle
        if vehicle and vehicle.position:
            # Train identifier
            train_id = vehicle.vehicle.id if vehicle.vehicle.id else entity.id
            train_label = vehicle.vehicle.label if vehicle.vehicle.label else train_id

            # Trip/route info
            route_id = ""
            trip_id = ""
            direction_id = -1
            if vehicle.trip:
                route_id = vehicle.trip.route_id if vehicle.trip.route_id else ""
                trip_id = vehicle.trip.trip_id if vehicle.trip.trip_id else ""
                direction_id = vehicle.trip.direction_id

            # Stop and status
            stop_id = vehicle.stop_id if vehicle.stop_id else ""
            status_map = {0: "INCOMING_AT", 1: "STOPPED_AT", 2: "IN_TRANSIT_TO"}
            current_status = status_map.get(vehicle.current_status, "UNKNOWN")

            # Position data
            speed = vehicle.position.speed if vehicle.position.speed else 0.0
            bearing = vehicle.position.bearing if vehicle.position.bearing else 0.0

            trains.append({
                "train_id": train_id,
                "train_label": train_label,
                "train_lat": float(vehicle.position.latitude),
                "train_long": float(vehicle.position.longitude),
                "train_speed": float(speed),
                "train_bearing": float(bearing),
                "route_id": route_id,
                "trip_id": trip_id,
                "direction_id": int(direction_id),
                "stop_id": stop_id,
                "current_status": current_status,
                "timestamp": datetime.now(sydney_tz).strftime("%Y-%m-%d %H:%M:%S")
            })

    return pd.DataFrame(trains)
```

```python
def dataframe_to_json_messages(df):
    """Convert DataFrame rows to JSON-encoded ServiceBus messages."""
    messages = []
    for _, row in df.iterrows():
        message_body = json.dumps(row.to_dict())
        messages.append(ServiceBusMessage(message_body))
    return messages
```

```python
def send_to_eventstream(connection_string, messages):
    """Send messages to Eventstream via Custom Endpoint (Event Hub compatible)."""
    # Parse EntityPath from connection string
    entity_path = None
    for param in connection_string.split(';'):
        if param.startswith('EntityPath='):
            entity_path = param.split('=')[1]
            break
    if not entity_path:
        raise ValueError("EntityPath not found in connection string")

    client = ServiceBusClient.from_connection_string(connection_string)
    try:
        with client.get_queue_sender(entity_path) as sender:
            sender.send_messages(messages)
    finally:
        client.close()
```

```python
def monitor_train_positions(api_key, connection_string, interval_seconds=15):
    """Continuously poll train positions and push to Eventstream."""
    sydney_tz = pytz.timezone("Australia/Sydney")
    print(f"Starting Sydney Trains position monitoring (interval: {interval_seconds}s)")
    print("Using v2 API endpoint. Press Ctrl+C to stop.")

    while True:
        try:
            df = get_train_positions(api_key)

            if df is not None and not df.empty:
                messages = dataframe_to_json_messages(df)
                send_to_eventstream(connection_string, messages)
                # Show route summary
                route_counts = df['route_id'].value_counts().head(5).to_dict()
                summary = ', '.join([f"{k}:{v}" for k,v in route_counts.items()])
                print(f"{datetime.now(sydney_tz).strftime('%H:%M:%S')}: Sent {len(messages)} trains | Top lines: {summary}")
            else:
                print(f"{datetime.now(sydney_tz).strftime('%H:%M:%S')}: No train data available")

            time.sleep(interval_seconds)

        except KeyboardInterrupt:
            print("Monitoring stopped.")
            break
        except Exception as e:
            print(f"Error: {e}")
            time.sleep(interval_seconds)
```
