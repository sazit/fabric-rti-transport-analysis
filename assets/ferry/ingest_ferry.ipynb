## Real-Time Ferry Position Monitor

This notebook connects to the Transport NSW Sydney Ferries API and feeds incoming position data into an Eventstream.

Based on the approach by [Francesco Fava](https://medium.com/@francescogiorgio.fava/real-time-ferry-tracking-in-sydney-harbour-with-microsoft-fabric-07bb2784ca50).

```python
import requests
import pandas as pd
from google.transit import gtfs_realtime_pb2
from azure.servicebus import ServiceBusClient, ServiceBusMessage
import json
import time
from datetime import datetime
import pytz
```

```python
# Transport NSW API key - replace with your key
myapikey = ""

# Eventstream Custom Endpoint connection string - get from Stream_Ferry_Loc
myconnectionstring = ""
```

```python
def get_ferry_positions(api_key):
    """Fetch real-time ferry positions from Transport NSW Sydney Ferries API."""
    url = "https://api.transport.nsw.gov.au/v1/gtfs/vehiclepos/sydneyferries"
    headers = {"Authorization": f"apikey {api_key}"}

    response = requests.get(url, headers=headers)
    response.raise_for_status()

    feed = gtfs_realtime_pb2.FeedMessage()
    feed.ParseFromString(response.content)

    ferries = []
    sydney_tz = pytz.timezone("Australia/Sydney")

    for entity in feed.entity:
        vehicle = entity.vehicle
        if vehicle and vehicle.position:
            ferry_name = vehicle.vehicle.label or vehicle.vehicle.id or entity.id
            destination = vehicle.trip.route_id if vehicle.trip and vehicle.trip.route_id else "Unknown"
            ferries.append({
                "ferry_name": ferry_name,
                "ferry_lat": vehicle.position.latitude,
                "ferry_long": vehicle.position.longitude,
                "ferry_destination": destination,
                "timestamp": datetime.now(sydney_tz).strftime("%Y-%m-%d %H:%M:%S")
            })

    return pd.DataFrame(ferries)
```

```python
def dataframe_to_json_messages(df):
    """Convert DataFrame rows to JSON-encoded ServiceBus messages."""
    messages = []
    for _, row in df.iterrows():
        message_body = json.dumps(row.to_dict())
        messages.append(ServiceBusMessage(message_body))
    return messages
```

```python
def send_to_eventstream(connection_string, messages):
    """Send messages to Eventstream via Custom Endpoint (Event Hub compatible)."""
    client = ServiceBusClient.from_connection_string(connection_string)
    sender = client.get_queue_sender(queue_name="ehub")
    with sender:
        for msg in messages:
            sender.send_messages(msg)
    client.close()
```

```python
def monitor_ferry_positions(api_key, connection_string, interval_seconds=15):
    """Continuously poll ferry positions and push to Eventstream."""
    print("Starting ferry position monitor...")
    print(f"Polling every {interval_seconds} seconds. Press Ctrl+C to stop.")

    iteration = 0
    while True:
        try:
            iteration += 1
            df = get_ferry_positions(api_key)

            if not df.empty:
                messages = dataframe_to_json_messages(df)
                send_to_eventstream(connection_string, messages)
                print(f"[{datetime.now().strftime('%H:%M:%S')}] Iteration {iteration}: Sent {len(messages)} ferry positions")
            else:
                print(f"[{datetime.now().strftime('%H:%M:%S')}] Iteration {iteration}: No ferry data returned")

            time.sleep(interval_seconds)

        except KeyboardInterrupt:
            print(f"\nStopped after {iteration} iterations.")
            break
        except Exception as e:
            print(f"Error: {e}")
            time.sleep(interval_seconds)
```
